```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = {
    'Annual Income': [15, 16, 17, 18, 19, 20, 60, 62, 65, 70, 72, 75],
    'Spending Score': [39, 81, 6, 77, 40, 76, 55, 60, 58, 80, 82, 90]
}

df = pd.DataFrame(data)
df
```

```python
X = df[['Annual Income', 'Spending Score']]
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
plt.scatter(df['Annual Income'], df['Spending Score'], c=df['Cluster'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Customer Segmentation using K-Means')
plt.show()
```

```python
## Conclusion

This project uses K-Means clustering to segment customers based on their
annual income and spending score. The model successfully groups customers
into distinct clusters, helping businesses understand customer behavior.
```
